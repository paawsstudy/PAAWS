```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
accelerometer_data = ""
IMU_data = ""
AGD_data = ""
```

```python
# function to read file
def read_data(filename, agd = False):
    sampling_rate = 1
    start_date = None
    start_time = None
    with open(filename) as f:
        line = f.readline()
        parsed = line.split()
        for i in range(len(parsed)):
            if parsed[i] == 'Hz':
                sampling_rate = int(parsed[i - 1])
                break
        f.readline()
        start_time = f.readline().split()[-1]
        start_date = f.readline().split()[-1]
    start = datetime.strptime(start_date + ' ' + start_time, '%m/%d/%Y %H:%M:%S')
    step = timedelta(seconds=1 / sampling_rate)
    if agd:
        step = timedelta(seconds = 1)
    df = pd.read_csv(filename,skiprows=10,header=0)
    df['Timestamp'] = [start + i* step for i in range(len(df))]
    return df
```

```python
%%time
accelerometer_df = read_data(accelerometer_data)
accelerometer_df
```

```python
%%time
IMU_df = read_data(IMU_data)
IMU_df
```

```python
%%time
agd_df = read_data(AGD_data)
agd_df
```
